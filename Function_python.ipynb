1. Difference between a Function and a Method in Python
- Function: A block of reusable code that is not tied to any object. It is defined using the def keyword.

        def greet(name):
          return "Hello " + name
 Method: A function that is associated with an object and is called on that object (usually using dot . notation).

           name = "Sanjana"
            print(name.upper())  # upper() is a method of the str object

2. Function Arguments vs Parameters

 Arguments: The actual values passed to the function when calling it.

 result = add(3, 4)  # 3 and 4 are arguments

4. Purpose of the return Statement
- Ends function execution and sends a value back to the caller.

Optional. If not used, the function returns None by default.

        def square(x):
        return x * x

5. Iterators vs Iterables
- Iterable: Any object that can return an iterator (e.g.,   list, tuple, string).

Iterator: An object with a __next__() method, which returns the next item in the sequence.

Example:

lst = [1, 2, 3]     # Iterable
it = iter(lst)      # Iterator created from list
print(next(it))     # 1

6. Generators
- A special type of iterator defined using the yield keyword.

Generates values on the fly without storing them in memory.

        def count_up_to(n):
        i = 1
        while i <= n:
        yield i
        i += 1

7. Advantages of Generators
Memory efficient: Don’t store entire sequences in memory.

Lazy evaluation: Values are generated only when needed.

Can represent infinite sequences.

8. Lambda Functions
Anonymous, one-line functions.

Syntax: lambda arguments: expression

Used when a small function is needed temporarily.

      square = lambda x: x * x
      print(square(5))  # 25

9. Purpose and Usage of map()
- Applies a function to every item of an iterable and returns a map object.

      numbers = [1, 2, 3]
      squares = map(lambda x: x ** 2, numbers)
      print(list(squares))  # [1, 4, 9]

10. map() vs filter() vs reduce()
- Function	Purpose	Example
map()	Applies a function to each item	map(lambda x: x*2, [1, 2, 3]) → [2, 4, 6]
filter()	Filters items based on condition	filter(lambda x: x%2==0, [1, 2, 3]) → [2]
reduce()	Applies rolling computation to pairwise items	reduce(lambda x, y: x + y, [1, 2, 3, 4]) → 10

reduce() is in functools module.

```python
1. Sum of All Even Numbers in a List

def sum_even_numbers(numbers):
    return sum(num for num in numbers if num % 2 == 0)

# Example
print(sum_even_numbers([1, 2, 3, 4, 5, 6]))  # Output: 12

2. Reverse a String

def reverse_string(s):
    return s[::-1]

# Example
print(reverse_string("hello"))  # Output: "olleh"

3. Squares of Each Number in a List

def square_list(numbers):
    return [num ** 2 for num in numbers]

# Example
print(square_list([1, 2, 3, 4]))  # Output: [1, 4, 9, 16]

4. Check if a Number is Prime (1 to 200)

def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# Test primes from 1 to 200
primes = [i for i in range(1, 201) if is_prime(i)]
print(primes)

5. Iterator Class for Fibonacci Sequence

class Fibonacci:
    def __init__(self, n_terms):
        self.n = n_terms
        self.count = 0
        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.n:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return 0
        elif self.count == 1:
            self.count += 1
            return 1
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.b

# Example
for num in Fibonacci(10):
    print(num, end=" ")

6. Generator for Powers of 2

def powers_of_two(n):
    for i in range(n + 1):
        yield 2 ** i

# Example
for power in powers_of_two(5):
    print(power)  # Output: 1, 2, 4, 8, 16, 32

7. Generator to Read File Line by Line

def read_file_line_by_line(filepath):
    with open(filepath, 'r') as file:
        for line in file:
            yield line.strip()

# Example usage:
# for line in read_file_line_by_line('sample.txt'):
#     print(line)

8. Lambda to Sort Tuples by Second Element

data = [(1, 3), (2, 1), (4, 2)]
sorted_data = sorted(data, key=lambda x: x[1])
print(sorted_data)  # Output: [(2, 1), (4, 2), (1, 3)]

9. Convert Celsius to Fahrenheit using map()

celsius = [0, 20, 30, 100]
fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))
print(fahrenheit)  # Output: [32.0, 68.0, 86.0, 212.0]

10. Remove Vowels from a String using filter()

def remove_vowels(s):
    return ''.join(filter(lambda x: x.lower() not in 'aeiou', s))

# Example
print(remove_vowels("Hello World"))  # Output: "Hll Wrld"
```
